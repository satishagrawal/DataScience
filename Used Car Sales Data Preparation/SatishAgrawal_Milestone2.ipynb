```python
# Load required binaries
import numpy as np
import array as arr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the dataset
df = pd.read_csv("USA_cars_datasets.csv")
df.head(10)
```

# Replace headers

```python
df.columns = ['Index', 'SellPrice', 'Brand', 'Model', 'Year', 'TitleStatus', 'TotalMiles', 'ExteriorColor', 'VIN', 'Lot#', 'State', 'Country', 'Time']
df.head(10)
```

# Format data into a more readable format & Fix casing or inconsistent values

```python
df.Country.unique()
```

```python
df["Country"].replace(" usa", "USA", inplace=True)
df.Country.replace(' canada', 'Canada', inplace=True)
```

```python
df.head()
```

```python
df.Country.unique()
# both casings are fixed and also extra leading space is trimmed out
```

```python
df.State = df['State'].str.capitalize()
df.Brand = df['Brand'].str.capitalize()
df.Model = df['Model'].str.capitalize()
```

```python
# Add a new column for location which represents city and state together
df['Location'] = df['State'] + ', ' +df['Country']
df.head()
```

```python
df['Age'] = 2020 - df.Year
df.head()
```

# Find duplicates

```python
# key here in this data set is the VIN#. Trying to find if there are cars appearing multiple times in the data set
any(df.VIN.duplicated())
```

```python
# looks like there are duplicates. Lets identify them
dr = df[df.duplicated(['VIN'])]
dr
```

```python
print("Dataframe size with duplicates:", df.shape)
df = df.drop_duplicates(subset='VIN', keep='first') 
df.shape
# looks like 4 duplicate rows were dropped
```

# Identify outliers and bad data

```python
print("Any null VINs:", df.VIN.isnull().values.any())
print("Any null SellPrices:", df.SellPrice.isnull().values.any())
print("Any null Brands:", df.Brand.isnull().values.any())
print("Any null in dataframe:", df.isnull().values.any())
```
